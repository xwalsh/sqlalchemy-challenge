# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
measurement = Base.classes.measurement
station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Find the most recent date in the data set.
recent_date = session.query(measurement_table.date).order_by(measurement_table.date.desc()).first()
recent_date
```

```python
latest_date = dt.datetime.strptime(str(recent_date), "('%Y-%m-%d',)").date()
one_year_ago = latest_date - dt.timedelta(days=365)


precipitation_data = session.query(
    measurement_table.date, measurement_table.prcp).filter(measurement_table.date >= one_year_ago).all()


precipitation_df = pd.DataFrame(precipitation_data, columns=['Date', 'Precipitation'])


precipitation_df.sort_values(by='Date', inplace=True)

precipitation_df.plot(x='Date', y='Precipitation', ylabel='Precipitation', legend=True)
plt.xticks(rotation=45)
plt.title("12 Months Precipitation")
plt.tight_layout()
plt.show()
```
